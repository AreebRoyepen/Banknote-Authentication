# Support vector machine implementation

### Our task is to predict whether a bank currency note is authentic or not based upon four attributes of the note i.e. skewness of the wavelet transformed image, variance of the image, entropy of the image, and curtosis of the image. This is a binary classification problem and we will use SVM algorithm to solve this problem

# Kernel SVM

# Implementing Kernel SVM with Scikit-Learn

### Importing Libraries

```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd 
```

### Importing the Dataset


```python
data_train = np.genfromtxt('group21_train.csv',delimiter=' ')
data_ = np.genfromtxt('group21_cv.csv', delimiter = ' ')
```

## Train Test Split

```python
X_train = data_train[:, 0:4]
y_train = data_train[:,4]
X_test = data_[:, 0:4]
y_test = data_[:,4]
```

## 2. Gaussian Kernel

```python
from sklearn.svm import SVC 
from sklearn.metrics import classification_report, confusion_matrix, f1_score

gammas = [2**-3, 2**-1, 2**1, 2**3, 2**5, 2**7, 2**9, 2**11]
c_vals = [2**-11, 2**-9, 2**-7, 2**-5, 2**-3, 2**-1, 2**1, 2**3]

gammas2 = []
c_vals2 = []
f1 = []

#Finding and displaying gamma and c values with classification report and confusion matrix
for g in gammas:
    for c in c_vals:
        gammas2.append(g)
        c_vals2.append(c)
        svclassifier = SVC(kernel='rbf', gamma = g, C = c) 
        svclassifier.fit(X_train, y_train) 
        
        y_pred = svclassifier.predict(X_test)  
         
        print(confusion_matrix(y_test, y_pred))  
        print(classification_report(y_test, y_pred))
        f1.append(f1_score(y_test, y_pred, average='macro'))
        print("Value of gamma: ", g) #value of gamma
        print("Value of c: ", c ) #value of c
        print("*****************************************************")
        
 
```

### It can be seen that there are several gamma and c value values which produce the maximum f1 score

```python
#Creating arrays of c, gamma and f1 score to plot graph
gammas2 = [-3, -1, 1, 3, 5, 7, 9, 11]
c_vals2 = [-11, -9, -7, -5, -3, -1, 1, 3]
gammas3 = []
c_vals3 = []

for g in gammas2:
    for c in c_vals2:
        gammas3.append(g)
        c_vals3.append(c)
```

```python
#Plotting graph of (gamma,c,f1 score)
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
fig.set_size_inches(18.5, 10.5)
ax = fig.add_subplot(111, projection='3d')

x = gammas3
y = c_vals3
z = f1



ax.scatter(x, y, z, c='r', marker='o')

ax.set_xlabel('Gamma (2^)')
ax.set_ylabel('C (2^)')
ax.set_zlabel('F1 Score')

plt.show()
```

### The graph makes it clear that there are several combinations of gamma and c which produce the maximum f1 score, therefore any combination of gamma and c which produces this maximum f1 score can be chosen when creating the classifier.

### Prediction and Evaluation

```python
#creating classifier using optimized gamma and c values
svclassifier = SVC(kernel='rbf', gamma = 2**-3, C = 2)
svclassifier.fit(X_train, y_train) 
```

```python
y_pred = svclassifier.predict(X_test) 
```

```python
print(confusion_matrix(y_test, y_pred))  
print(classification_report(y_test, y_pred))
```
